```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import LabelEncoder
import itertools
from sklearn.preprocessing import minmax_scale

plt.style.use('default')
```

## Подготовка

### Преобразование датасета

```python
# читаем датасет, преобразуем даты
df = pd.read_csv('dataset.csv')
df['dt'] = pd.to_datetime(df['dt'])
df['user_registered_time'] = pd.to_datetime(df['user_registered_time'])


# кодируем строки
user_sex = {1: 'Мужчина', 2: 'Женщина'}
marital_status = {
    0: 'Не выбрано', 
    1: 'Не состоит в браке',
    2: 'Встречается', 
    3: 'Помолвлен', 
    4: 'Состоит в браке', 
    5: 'Все сложно', 
    6: 'В поиске', 
    7: 'Влюблен',
    8: 'В гражданском браке'
}
df['user_sex'] = df['user_sex'].apply(lambda x: user_sex[x])
df['user_marital_status'] = df['user_marital_status'].apply(lambda x: marital_status[x])
```

```python
# Таблица стран пользователей
user_country = df.groupby('user_id')['auditory_country_id']
user_country = user_country.apply(lambda x: x.value_counts().sort_values().index[-1])

# Таблица полов пользователей
user_sex = df.groupby('user_id')['user_sex']
user_sex = user_sex.apply(lambda x: x.value_counts().sort_values().index[-1])

# Таблица количества посещений пользователей за три месяца
user_visits = df.groupby('user_id')['dt'].count()

# таблица регистраций пользователей
user_register = df.groupby('user_id')['user_registered_time'].apply(lambda x: x.values[0])
```

```python
# Создаем таблицу со средними
mean_df = df.groupby('user_id').mean()
mean_df = mean_df.merge(user_country, left_index=True, right_index=True)
mean_df = mean_df.merge(user_sex, left_index=True, right_index=True)
mean_df = mean_df.merge(user_visits, left_index=True, right_index=True)
mean_df = mean_df.merge(user_register, left_index=True, right_index=True)
mean_df.rename(columns={'dt': 'visits'}, inplace=True)

other_apps = ['facebook', 'facebook_messenger', 'whatsapp', 
              'telegram', 'instagram', 'skype', 'viber', 'tiktok', 'ok']
for app in other_apps:
    mean_df[f'has_app_{app}'] = mean_df[f'has_app_{app}'].astype(int)
```

```python
# Разбиение на группы активности
activity_labels = ['not-user', 'visiter', 'user', 'active', 'super-active']
# activity_types = ['message_activity', 'calls_activity', 'comments_activity']

mean_df['message_activity'] = pd.cut(mean_df['messaging_total'], 
                                     bins=[-1, 0, 1, 10, 100, 1000], 
                                     labels=activity_labels)
mean_df['calls_activity'] = pd.cut(mean_df['calls'], 
                                   bins=[-1, 0, 0.05, 0.2, 2, mean_df['calls'].max() + 1], 
                                   labels=activity_labels)
mean_df['comments_activity'] = pd.cut(mean_df['comments_created'], 
                                   bins=[-1, 0, 0.1, 0.5, 4, mean_df['calls'].max() + 1], 
                                   labels=activity_labels)
```

### Полезные функции

```python
def show_activity_kde(activity_type):
    """
    Строит график использования сервиса по уровням активности
    
    :param activity_type: {'message_activity', 'calls_activity', 'comments_activity'}
    """
    
    fig, axs = plt.subplots(2, 2, figsize=(10, 6))
    column = {'message_activity': 'messaging_total', 
              'calls_activity': 'calls', 
              'comments_activity': 'comments_created'}[activity_type]
    for i, ac in enumerate(activity_labels[1:]):
        sns.kdeplot(data=mean_df[mean_df[activity_type] == ac], x=column, 
                 hue='auditory_country_id', ax=axs[i // 2, i % 2], common_norm=False)
        

    for i, title in enumerate(activity_labels[1:]):
        axs[i // 2, i % 2].set_title(f'уровень активности - {title}')

    x_label = {'message_activity': 'Сообщений в день', 
               'calls_activity': 'Звонков в день', 
               'comments_activity': 'Комментариев в день'}[activity_type]
    for i in range(4):
        axs[i // 2, i % 2].set_xlabel(x_label)
    
    suptitle = {'message_activity': 'Среднее ежедневное количество сообщений',
                'calls_activity': 'Среднее ежедневное количество звонков',
                'comments_activity': 'Среднее ежедневное количество комментариев'}[activity_type]
    fig.suptitle(suptitle, fontsize=15)
    
    fig.tight_layout()
```

```python
def show_activity_table(activity_1, activity_2):
    """
    Строит таблицу пересечений уровней активностей между двумя видами активностей 
    """
    
    fig, axs = plt.subplots(1, 2, figsize=(7, 4))
    
    
    for i, country in enumerate(['Кыргызстан', 'Латвия']):
        activity_counts = mean_df[mean_df['auditory_country_id'] == country][[activity_1, activity_2]].value_counts()
        activity_counts = activity_counts.reset_index().pivot_table(index=[activity_1], columns=[activity_2], values=0)
        activity_counts = activity_counts.drop('not-user', axis=1).drop('not-user', axis=0).values
        cmap = [plt.cm.Blues, plt.cm.Oranges][i]
        axs[i].imshow(activity_counts, cmap=cmap)
        axs[i].set_title(country)
        
        perc = activity_counts / np.nansum(activity_counts) * 100
        perc = np.around(perc, decimals=1)
        for a in range(len(activity_counts)):
            for b in range(len(activity_counts[a])):
                color = 'white' if activity_counts[a, b] > np.nanmax(activity_counts) / 2 else 'black'
                axs[i].text(b, a, f'{perc[a, b]}%', color=color, ha='center')
                
        axs[i].set_xticks([0,1,2,3])
        axs[i].set_xticklabels(activity_labels[1:])
        axs[i].set_yticks([0,1,2,3])
        axs[i].set_yticklabels(activity_labels[1:], rotation=90)
    
    axs[0].set_xlabel(activity_1)
    axs[1].set_xlabel(activity_1)
    axs[0].set_ylabel(activity_2)
    
    fig.suptitle(f'Комбинации между активностями {activity_1} и {activity_2}', fontsize=15)    
    fig.tight_layout()
```

```python
def show_activity_distribution(activity_type):
    """
    Строит график степеней активности пользователей по типу активности
    """
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    counts_activity = mean_df.groupby('auditory_country_id')[activity_type].value_counts()

    for i, country in enumerate(['Кыргызстан', 'Латвия']):
        color = ['C0', 'C1'][i]
        axs[i].bar(x=counts_activity[country].index, height=counts_activity[country], color=color)
        axs[i].set_ylim([0, counts_activity[country].max() * 1.15])
        perc = np.around(counts_activity[country] / counts_activity[country].sum() * 100, decimals=1)
        perc_margin = counts_activity[country].max() * 0.15 / 3

        for j, v in counts_activity[country].items():
            axs[i].text(j, v + perc_margin, f'{perc[j]}%', ha='center')
        axs[i].set_title(country)
        axs[i].set_xlabel('Тип активности пользователя')
        axs[i].set_ylabel('Количество пользователей')

    fig.suptitle(activity_type);
```

### Демография

```python
sex_count = mean_df.groupby('auditory_country_id')['user_sex'].value_counts()
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
for i, country in enumerate(['Кыргызстан', 'Латвия']):
    color = ['C0', 'C1'][i]
    axs[i].bar(x=sex_count[country].index, height=sex_count[country].values, color=color)
    for ind, val in sex_count[country].items():
        perc = np.around(sex_count[country] / sex_count[country].sum() * 100, decimals=2)
        axs[i].text(ind, val/2, f'{perc[ind]}%', ha='center', va='center', color='white', fontsize=13)
    axs[i].set_title(country)
fig.suptitle('Распределение полов', fontsize=15)
fig.tight_layout();
```

```python
# Добавлено после собеседования:
# Проверка на статистическую значимость различия долей полов в каждой стране

sex_contingency_table = sex_count.reset_index(level=0).pivot(columns='auditory_country_id', values='user_sex').T
sex_contingency_table
chi2, p, dof, ex = scipy.stats.chi2_contingency(sex_contingency_table)
print(f'p_value: {p}\nchi2: {chi2}')
```

В Латвии пользователей мужчин не намного больше пользователей женщин. Это объясняется демографией страны (0.85 male(s)/female ~ 46/53)

Среди пользователей Кыргызстана намного больше женщин, и демография страны этого не объясняет, ведь среди населения Кыргызстана соотношение полов примерно одинаково (0.96 male(s)/female).

# Критерий активности

Активность пользователя выражается четырьмя параметрами:

- насколько активно ведет переписку
- как часто использует звонки
- как часто оставляет комментарии

Чем выше эти параметры - тем активнее пользователь. Поэтому, можно применить метрику "Расстояние активности".

$$AD = \sqrt{{coments}^2 + {messages}^2 + {calls}^2}$$

где 
* **comments** - среднее число комментариев, оставляемых пользователем в день
* **messages** - среднее число сообщений, отправляемых пользователем в день 
* **calls** - среднее число звонков, совершаемых пользователем в день

Так как сообщения используются чаще всего, а звонки реже, стоит привести параметры к одной шкале измерения. Иначе комментарии и звонки не будут особо влиять на метрику.

Также, до того как приводить параметры к одной шкале, их нужно логарифмировать, чтобы метрика не уходила совсем в ноль на малоактивных пользователях.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_params = np.log(mean_df[['messaging_total', 'calls', 'comments_created']] + 1)
scaled_params = scaler.fit_transform(scaled_params)
activity = np.sqrt(np.square(scaled_params).sum(axis=1))
mean_df['activity'] = np.log(activity + 1)
```

```python
sns.kdeplot(data=mean_df[mean_df['activity'] > 0], x='activity', 
            hue='auditory_country_id', common_norm=False, color=['C1', 'C0'])
plt.title('Активность пользователей по странам');
```

Очевидно, Латыши немного активнее жителей Кыргызстана. Но доля малоактивных пользователей сообщества все равно велика. 

Наша метрика это показывает в совокупности со всеми сервисами сообщества.

# Активность

Активность пользователя может измеряться в его

- Объему передаваемой информации на одного пользователя (messages_per_user)
- Инициативности (initiative)
- Ответной реакции (responsiveness)
- Широтой круга общения (sociability)

Далее подробнее об этих показателях:

### messages_per_user

#### Формула

$$messages\_per\_user = \frac{messages\_to\_users}{unique\_users} + messages\_to\_chats$$

где 
- $messages\_to\_users$ - количество сообщений, отправленных пользователям
- $unique\_users$ - количество пользователей, которым были отправлены сообщения
- $messages\_to\_chats$ - количество сообщений, отправленных в чат



### initiative

#### Формула

$$initiative = \frac{messages\_init\_users}{unique\_users}$$

где 
- $messages\_init\_users$ - количество пользователей, которым пользователь отправил сообщение первым
- $unique\_users$ - количество пользователей, которым были отправлены сообщения


### responsiveness

#### Формула

$$responsiveness = \frac{messages\_init\_with\_reply}{messages\_init\_users}$$

где 
- $messages\_init\_with\_reply$ - количество пользователей, ответивших на сообщение пользователя 
- $messages\_init\_users$ - количество пользователей, которым пользователь отправил сообщение первым


### sociability

#### Формула

$$sociability = chat\_users + unique\_users$$

где 
- $unique\_users$ - количество пользователей, которым были отправлены сообщения в личных сообщениях
- $chat\_users$ - количество пользователей, ответивших на сообщение пользователя в чате


----


## Метрика активности в сообщениях (Message Activity)

$$Message\_Activity = initiative * responsiveness + log_2(sociability + 1) * np.log_{10}(messages\_per\_user + 1)$$

### Логика

Показатели *initiative* и *responsiveness* варьируются в пределах от 0 до 1.

||Низкая инициативность|Высокая инициативность|
|--|--|--|
|**Низкая ответная реакция**|*редко* начинает диалог, *редко* получает ответ|*часто* начинает диалог, *редко* получает ответ|
|**Высокая ответная реакция**|*редко* начинает диалог, *часто* получает ответ|*часто* начинает диалог, *часто* получает ответ|

---

*sociability* и *messages_per_user* лежат в пределе $[0, +\infty)$

по сути $log_2(sociability + 1) * np.log_{10}(messages\_per\_user + 1)$ обозначает количество информации, переданной пользователем.


```python
messages_per_user = (df['messaging_total_to_users'] / df['messaging_to_uniq_users']).replace(np.nan, 0)
initiative = (df['messaging_inited_direct_users'] / df['messaging_to_uniq_users']).replace(np.nan, 0)
responsiveness = (df['messaging_inited_direct_with_reply'] / df['messaging_inited_direct_users']).replace(np.nan, 0)
sociability = df['messaging_inited_chat_users'] + df['messaging_to_uniq_users']
```

```python
fig, axs = plt.subplots(2, 2)

sns.histplot(initiative[(initiative > 0) & (initiative <= 1)], bins=10, ax=axs[0, 0])
axs[0, 0].set_title('initiative (0+)')

sns.histplot(responsiveness, bins=10, ax=axs[0, 1])
axs[0, 1].set_title('responsiveness')

sns.histplot(np.log10(messages_per_user[messages_per_user > 0] + 1), bins=20, ax=axs[1, 0])
axs[1, 0].set_title('messages_per_user (0+)')

sns.histplot(np.log2(sociability + 1), bins=20, ax=axs[1, 1])
axs[1, 1].set_title('sociability')

fig.tight_layout()
```

```python
activity = initiative * responsiveness + np.log2(sociability + 1) * np.log10(messages_per_user + 1)
df['activity_new'] = activity
```

```python
sns.histplot(data=df[df['activity_new'] > 0], x='activity_new', hue='auditory_country_id',
             bins=25, log_scale=True, common_norm=False, stat='density')
med = df[df['activity_new'] > 0].groupby('auditory_country_id')['activity_new'].median()
plt.axvline(med['Кыргызстан'], linestyle='dashed', color='C0', 
            label=f"Кыргызстан: (med: {np.around(med['Кыргызстан'], 3)})")
plt.axvline(med['Латвия'], linestyle='dashed', color='C1', 
            label=f"Латвия: (med: {np.around(med['Латвия'], 3)})")
plt.legend()
plt.title('Распределение метрики активности по странам\nсреди активных пользователей (>0)');
```
